### **Sentiment Analysis**

```python
from transformers import pipeline
classifier=pipeline("sentiment-analysis")
```

```python
task=("This Nlp and Hugging Face course content is very good!")
classifier(task)
```

```python
new_task=["This Nlp and Hugging Face course content is very good!", "But Balancing work and learning has been quite exhausting lately"]
classifier(new_task)
```

```python
classifier(
    ["This Nlp and Hugging Face course content is very good!", "But Balancing work and learning has been quite exhausting lately"]
)
```

```python
results=classifier(new_task)
for result in results:
  print(f"Label: {result['label']}---> Score: {result['score']}")
```

### **Zero-Shot Classification**

#### **Example No: 1**

```python
from transformers import pipeline
zeroshot_classifier=pipeline("zero-shot-classification")
tasks=zeroshot_classifier(
    "This is the Nlp and Hugging Face Course",
    candidate_labels =["Education", "Teaching", "Politics","literacy"]
)
print(tasks)
```

```python
for label,score in zip(tasks['labels'], tasks['scores']):
  print(label,'-->',score)
```

```python
for task in tasks:
  print(f"Label: {tasks['labels']}---> Score: {tasks['scores']}")
```

#### **Example No: 2**

```python
zeroshot_tasks=zeroshot_classifier(
    "This is the Nlp course and it make my concepts very strong related to NLP",
    candidate_labels=["literacy", "Education", "Politics", "Teaching"]
)
for label, score in zip(zeroshot_tasks['labels'], zeroshot_tasks['scores']):
  print("Label:", label, '--->', "Score:", score)
  # print(f"Label: {zeroshot_tasks['labels']}---> Score: {zeroshot_tasks['scores']}")
```

#### **max_length parameter**

```python
from transformers import pipeline

# Create the text generation pipeline
text_generator = pipeline("text-generation", max_length=50)

# Generate the output
output = text_generator(
    "I'm learning NLP Course today and I have learned a lot from this course. Now I have to do",
    max_length=50
)

# Print each generated text cleanly
for i, result in enumerate(output, 1):
    print(f"Output {i}:\n{result['generated_text']}\n")
```

#### **max_length parameter, num_return_sequences**

```python
from transformers import pipeline

# Create the text generation pipeline
text_generator = pipeline("text-generation", model="gpt2")

# Generate 5 different sequences
output = text_generator(
    "I'm learning NLP Course today and I have learned a lot from this course. Now I have to do",
    max_length=50,
    num_return_sequences=5
)

# Print outputs 1 to 5
for i, result in enumerate(output, 1):
    print(f"Output {i}:\n{result['generated_text']}\n")
```

```python
text_generator = pipeline("text-generation", model="distilgpt2")
output = text_generator("I'm learning NLP Course today and I have learned a lot from this course. Now I have to do", max_length=50)
print(output)
```

### **Named Entity Recognition (NER)**

```python
ner_classifier=pipeline("ner", model="dslim/bert-base-NER")
ner_text="I'm doing Datascience and Ai and I have just completed by Bachelors and now i'm creating various projects on GitHub and kaggle"
print(ner_classifier(ner_text))
```

```python
ner_classifier=pipeline("ner", model="dslim/bert-base-NER", grouped_entities=True)
ner_text="I'm doing Datascience and Ai and I have just completed by Bachelors and now i'm creating various projects on GitHub and kaggle"
print(ner_classifier(ner_text))
```

### **Question Answering**

```python
from transformers import pipeline
qa_classifier = pipeline("question-answering", model="PlanTL-GOB-ES/roberta-base-bne-sqac")
qa= qa_classifier(
    context="I'm doing Datascience and Ai and I have just completed by Bachelors and now i'm creating various projects on GitHub and kaggle",
    question="What I'm doing nowadays"
)
print(qa)
```

```python
from transformers import pipeline
qa_classifier = pipeline("question-answering", model="PlanTL-GOB-ES/roberta-base-bne-sqac")
qa= qa_classifier(
    context="I'm doing Datascience and Ai and I have just completed by Bachelors and now i'm creating various projects on GitHub and kaggle",
    question="What I've studied?"
)
print(qa)
```
